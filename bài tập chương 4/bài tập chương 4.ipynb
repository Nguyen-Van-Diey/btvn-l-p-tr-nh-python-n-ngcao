```python
# bài 4.1

import numpy as np
import pandas as pd

mang_np = np.array([10, 20, 30, 40, 50])
chuoi_tu_np = pd.Series(mang_np)
print("Chuỗi pandas tạo từ numpy:")
print(chuoi_tu_np)
print("Kiểu dữ liệu:", type(chuoi_tu_np))
print()
tu_dien_diem = {
    "An": 7.5,
    "Bình": 8.0,
    "Chi": 9.0
}
chuoi_tu_dict = pd.Series(tu_dien_diem)
print("Chuỗi pandas tạo từ dict:")
print(chuoi_tu_dict)
print("Kiểu dữ liệu:", type(chuoi_tu_dict))
```

```python
# bài 4.2

import pandas as pd

ds_chuoi = ["Ngô,20,7.5","Bắp,21,8.0","Chuối,19,9.0","Đào,22,6.8"]
ds_hang = []
for dong in ds_chuoi:
    ho_ten, tuoi, diem = dong.split(",")
    ds_hang.append([ho_ten, tuoi, diem])
bang_sinh_vien = pd.DataFrame(ds_hang, columns=["HoTen", "Tuoi", "Diem"])
print("DataFrame ban đầu:")
print(bang_sinh_vien)
print(bang_sinh_vien.dtypes)  
print()
bang_sinh_vien["Tuoi"] = bang_sinh_vien["Tuoi"].astype(int)
bang_sinh_vien["Diem"] = bang_sinh_vien["Diem"].astype(float)
print("DataFrame sau khi ép kiểu:")
print(bang_sinh_vien)
print("\nKiểu dữ liệu từng cột:")
print(bang_sinh_vien.dtypes)
```

```python
import pandas as pd

chuoi_diem_toan = pd.Series([7.5, 8.0, 6.5, 9.0], index=["An", "Bình", "Chi", "Dung"])
chuoi_diem_ly   = pd.Series([8.0, 7.0, 7.5, 8.5], index=["An", "Bình", "Chi", "Dung"])
print("Chuỗi điểm Toán:")
print(chuoi_diem_toan)
print("\nChuỗi điểm Lý:")
print(chuoi_diem_ly)
bang_diem = pd.concat([chuoi_diem_toan, chuoi_diem_ly], axis=1)
bang_diem.columns = ["Toan", "Ly"]
print("\nBảng điểm ban đầu:")
print(bang_diem)
bang_diem_sorted_index = bang_diem.sort_index()
print("\nBảng điểm sắp xếp theo tên (index):")
print(bang_diem_sorted_index)
bang_diem_sorted_toan = bang_diem.sort_values(by="Toan", ascending=False)
print("\nBảng điểm sắp xếp theo điểm Toán (giảm dần):")
print(bang_diem_sorted_toan)
```
