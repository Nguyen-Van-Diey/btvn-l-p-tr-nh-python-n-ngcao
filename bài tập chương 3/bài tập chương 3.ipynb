```python
# bài 3.5
import numpy as np

# đọc chiều cao
ds_chieu_cao = []
with open("heights_1.txt", "r") as f:
    for dong in f:
        dong = dong.strip()
        if dong != "":
            ds_chieu_cao.append(float(dong))
# đọc cân nặng
ds_can_nang = []
with open("weights_1.txt", "r") as f:
    for dong in f:
        dong = dong.strip()
        if dong != "":
            ds_can_nang.append(float(dong))

print("Số phần tử chiều cao:", len(ds_chieu_cao))
print("Số phần tử cân nặng:", len(ds_can_nang))
# 1
mang_chieu_cao = np.array(ds_chieu_cao, dtype=float)   
mang_can_nang = np.array(ds_can_nang, dtype=float)     
# 2
he_so_inch_m = 0.0254
he_so_pound_kg = 0.453592
mang_chieu_cao_m = mang_chieu_cao * he_so_inch_m       
mang_can_nang_kg = mang_can_nang * he_so_pound_kg      
# 3
mang_bmi = mang_can_nang_kg / (mang_chieu_cao_m ** 2)  
# 4
if len(mang_can_nang_kg) > 50:
    print("Cân nặng (kg) tại index 50:", mang_can_nang_kg[50])
# 5
if len(mang_chieu_cao_m) > 110:
    mang_chieu_cao_m_100 = mang_chieu_cao_m[100:111]
    print("Chiều cao từ index 100 đến 110 (m):")
    print(mang_chieu_cao_m_100)
# 6
mat_na_bmi_thap = mang_bmi < 21
bmi_thap = mang_bmi[mat_na_bmi_thap]
print("Các giá trị BMI < 21:")
print(bmi_thap)
# 7
chieu_cao_tb = mang_chieu_cao_m.mean()
can_nang_tb = mang_can_nang_kg.mean()
print("Chiều cao trung bình (m):", chieu_cao_tb)
print("Cân nặng trung bình (kg):", can_nang_tb)
# 8.
chieu_cao_max = mang_chieu_cao_m.max()
can_nang_max = mang_can_nang_kg.max()
print("Chiều cao lớn nhất (m):", chieu_cao_max)
print("Cân nặng lớn nhất (kg):", can_nang_max)
# 9
chieu_cao_min = mang_chieu_cao_m.min()
can_nang_min = mang_can_nang_kg.min()
print("Chiều cao nhỏ nhất (m):", chieu_cao_min)
print("Cân nặng nhỏ nhất (kg):", can_nang_min)
```

```python
import numpy as np

# Đọc dữ liệu từ file baseball_2D.txt vào list
du_lieu_baseball = []
with open("baseball_2D.txt", "r") as f:
    for dong in f:
        dong = dong.strip()
        if dong == "":
            continue
        # mỗi dòng: chiều cao – cân nặng – tuổi – ...
        phan_tu = dong.split()              # tách theo khoảng trắng
        hang_so = [float(x) for x in phan_tu]
        du_lieu_baseball.append(hang_so)

# Câu 1: tạo numpy array 2D np_baseball
np_baseball = np.array(du_lieu_baseball)

print("Kiểu dữ liệu np_baseball:", np_baseball.dtype)
print("Kích thước np_baseball:", np_baseball.shape)   # (số_cầu_thủ, số_thuộc_tính)

# Câu 2: chiều cao của vận động viên thứ 50 (cột 1 → index 0)
chieu_cao_thu_50 = np_baseball[49, 0]      # vì index bắt đầu từ 0
print("Chiều cao VĐV thứ 50 (inch):", chieu_cao_thu_50)

# Câu 3: tạo np_weight từ cột 2 (index 1)
np_weight = np_baseball[:, 1]
print("Một vài cân nặng đầu tiên:", np_weight[:5])

# Câu 4: chiều cao của vận động viên thứ 124
chieu_cao_thu_124 = np_baseball[123, 0]
print("Chiều cao VĐV thứ 124 (inch):", chieu_cao_thu_124)

# Câu 5: chiều cao trung bình, cân nặng trung bình
np_height = np_baseball[:, 0]   # cột chiều cao

chieu_cao_tb = np_height.mean()
can_nang_tb = np_weight.mean()

print("Chiều cao trung bình (inch):", chieu_cao_tb)
print("Cân nặng trung bình (pounds):", can_nang_tb)

# (Bonus) tính hệ số tương quan giữa chiều cao và cân nặng
ma_tran_tuong_quan = np.corrcoef(np_height, np_weight)
he_so_tuong_quan = ma_tran_tuong_quan[0, 1]
print("Hệ số tương quan chiều cao – cân nặng:", he_so_tuong_quan)
```

```python
import numpy as np

ds_chieu_cao = []
with open("heights.txt", "r") as f:
    for dong in f:
        dong = dong.strip()
        if dong != "":
            ds_chieu_cao.append(float(dong))
ds_vi_tri = []
with open("positions.txt", "r") as f:
    for dong in f:
        dong = dong.strip()
        if dong != "":
            ds_vi_tri.append(dong)

print("Số dòng chiều cao:", len(ds_chieu_cao))
print("Số dòng vị trí    :", len(ds_vi_tri))

# 1a)
np_positions = np.array(ds_vi_tri, dtype="U5")
print("Kiểu dữ liệu np_positions:", np_positions.dtype)
# 1b) 
np_heights = np.array(ds_chieu_cao, dtype=float)
print("Kiểu dữ liệu np_heights:", np_heights.dtype)
# 2.
mat_na_gk = (np_positions == "GK")
chieu_cao_tb_gk = np_heights[mat_na_gk].mean()
print("Chiều cao trung bình GK:", chieu_cao_tb_gk)
# 3
mat_na_khong_gk = (np_positions != "GK")
chieu_cao_tb_khong_gk = np_heights[mat_na_khong_gk].mean()
print("Chiều cao trung bình các vị trí khác GK:", chieu_cao_tb_khong_gk)
# 4
kieu_player = np.dtype([("position", "U5"), ("height", "f4")])
players = np.array(list(zip(np_positions, np_heights)), dtype=kieu_player)
print("Kiểu dữ liệu players:", players.dtype)
print("Ví dụ vài phần tử đầu của players:")
print(players[:5])
# 5
players_sap_xep = np.sort(players, order="height")
player_thap_nhat = players_sap_xep[0]
player_cao_nhat = players_sap_xep[-1]
print("Cầuthủ thấp nhất:", player_thap_nhat)
print("Cầu thủ cao nhất :", player_cao_nhat)
print("Vị trí của cầu thủ thấp nhất:", player_thap_nhat["position"])
print("Vị trí của cầu thủ cao nhất :", player_cao_nhat["position"])
```
