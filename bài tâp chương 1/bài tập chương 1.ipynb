```python
# câu 1
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong
    def nhap(self):
        self.dai = float(input("Nhập chiều dài: "))
        self.rong = float(input("Nhập chiều rộng: "))
    def chu_vi(self):
        return 2 * (self.dai + self.rong)
    def dien_tich(self):
        return self.dai * self.rong
    def xuat(self):
        print(f"Hình chữ nhật:")
        print(f" - Chiều dài: {self.dai}")
        print(f" - Chiều rộng: {self.rong}")
        print(f" - Chu vi: {self.chu_vi()}")
        print(f" - Diện tích: {self.dien_tich()}")
```

```python
# câu 2
class ThiSinh:
    def __init__(self, ho_ten="", toan=0.0, ly=0.0, hoa=0.0):
        """Hàm tạo: khởi tạo thông tin thí sinh"""
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa
    def nhap(self):
        """Nhập thông tin thí sinh"""
        self.ho_ten = input("Nhập họ và tên thí sinh: ")
        self.toan = float(input("Nhập điểm Toán: "))
        self.ly = float(input("Nhập điểm Lý: "))
        self.hoa = float(input("Nhập điểm Hóa: "))
    def tong_diem(self):
        """Tính tổng điểm 3 môn"""
        return self.toan + self.ly + self.hoa
    def xuat(self):
        """In thông tin thí sinh"""
        print(f"Thí sinh: {self.ho_ten}")
        print(f"  Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}")
        print(f"  Tổng điểm: {self.tong_diem()}")
        print("-" * 40)
if __name__ == "__main__":
    ds = [] 
    n = int(input("Nhập số lượng thí sinh: "))

    for i in range(n):
        print(f"\n--- Nhập thí sinh thứ {i+1} ---")
        ts = ThiSinh()
        ts.nhap()
        ds.append(ts)
    trung_tuyen = [ts for ts in ds if ts.tong_diem() >= 20]
    trung_tuyen.sort(key=lambda x: x.tong_diem(), reverse=True)
    print("\n=== DANH SÁCH TRÚNG TUYỂN (>=20 điểm) ===")
    if not trung_tuyen:
        print("Không có thí sinh nào trúng tuyển.")
    else:
        for ts in trung_tuyen:
            ts.xuat()
```

```python
# câu 3
import math

class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau if mau != 0 else 1  
    def hop_le(self):
        return self.mau != 0
    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        while self.mau == 0:
            print("Mẫu số phải khác 0, vui lòng nhập lại!")
            self.mau = int(input("Nhập mẫu số: "))
    def rut_gon(self):
        ucln = math.gcd(self.tu, self.mau)
        self.tu //= ucln
        self.mau //= ucln
        if self.mau < 0:
            self.tu = -self.tu
            self.mau = -self.mau
    def xuat(self):
        self.rut_gon() 
        if self.mau == 1:
            print(f"Phân số: {self.tu}")
        else:
            print(f"Phân số: {self.tu}/{self.mau}")
if __name__ == "__main__":
    ps = PhanSo()
    ps.nhap()
    if ps.hop_le():
        ps.xuat()
    else:
        print("Phân số không hợp lệ (mẫu = 0)")
```

```python
# câu 4
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.data = []  
        print("Ngăn xếp được tạo.")
    def __del__(self):
        print("Ngăn xếp đã bị hủy.")
    def is_empty(self):
        return len(self.data) == 0
    def is_full(self):
        return len(self.data) == self.capacity
    def push(self, value):
        if self.is_full():
            print("Ngăn xếp đã đầy, không thể push!")
        else:
            self.data.append(float(value))
            print(f"Đã push {value}")
    def pop(self):
        if self.is_empty():
            print("Ngăn xếp rỗng, không thể pop!")
            return None
        else:
            value = self.data.pop()
            print(f"Đã pop {value}")
            return value
```

```python
# câu 9
class Polygon:
    def __init__(self, sides=[]):
        self.sides = sides
    def input_sides(self):
        n = int(input("Nhập số cạnh của đa giác: "))
        self.sides = []
        for i in range(n):
            self.sides.append(float(input(f"  Nhập cạnh {i+1}: ")))
    def perimeter(self):
        return sum(self.sides)
    def display(self):
        print("Các cạnh của đa giác:", self.sides)
        print("Chu vi đa giác:", self.perimeter())
class Parallelogram(Polygon):
    def __init__(self, a=0, b=0, h=0):
        super().__init__([a, b, a, b])
        self.a = a
        self.b = b
        self.h = h
    def area(self):
        return self.a * self.h
    def display(self):
        print("Hình bình hành:")
        print(f"  Cạnh a = {self.a}, cạnh b = {self.b}, chiều cao h = {self.h}")
        print("  Chu vi:", self.perimeter())
        print("  Diện tích:", self.area())
class Rectangle(Parallelogram):
    def __init__(self, a=0, b=0):
        super().__init__(a, b, b)
        self.a = a
        self.b = b
    def area(self):
        return self.a * self.b
    def display(self):
        print("Hình chữ nhật:")
        print(f"  Chiều dài = {self.a}, chiều rộng = {self.b}")
        print("  Chu vi:", self.perimeter())
        print("  Diện tích:", self.area())
class Square(Rectangle):
    def __init__(self, a=0):
        super().__init__(a, a)
        self.a = a
    def area(self):
        return self.a * self.a
    def display(self):
        print("Hình vuông:")
        print(f"  Cạnh = {self.a}")
        print("  Chu vi:", self.perimeter())
        print("  Diện tích:", self.area())
```

```python
# câu 10
import math
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    def input_point(self):
        self.x = float(input("Nhập hoành độ x: "))
        self.y = float(input("Nhập tung độ y: "))
    def display(self):
        print(f"Tọa độ điểm: ({self.x}, {self.y})")
class Ellipse(Point):
    def __init__(self, x=0, y=0, a=0, b=0):
        super().__init__(x, y)  
        self.a = a  
        self.b = b 
    def input_ellipse(self):
        print("Nhập tâm elip:")
        self.input_point()
        self.a = float(input("Nhập bán trục lớn a: "))
        self.b = float(input("Nhập bán trục nhỏ b: "))
    def area(self):
        return math.pi * self.a * self.b
    def display(self):
        print(f"Elip có tâm ({self.x}, {self.y}), a = {self.a}, b = {self.b}")
        print("Diện tích elip:", self.area())
class Circle(Ellipse):
    def __init__(self, x=0, y=0, r=0):
        super().__init__(x, y, r, r)  
        self.r = r
    def input_circle(self):
        print("Nhập tâm đường tròn:")
        self.input_point()
        self.r = float(input("Nhập bán kính r: "))
        self.a = self.r
        self.b = self.r
    def area(self):
        return math.pi * self.r * self.r
    def display(self):
        print(f"Đường tròn có tâm ({self.x}, {self.y}), bán kính r = {self.r}")
        print("Diện tích đường tròn:", self.area())
```

```python
# câu 11
import math

class Triangle:
    def __init__(self, a=0, b=0, c=0):
        self.a = a
        self.b = b
        self.c = c
    def input_triangle(self):
        self.a = float(input("Nhập cạnh a: "))
        self.b = float(input("Nhập cạnh b: "))
        self.c = float(input("Nhập cạnh c: "))
    def is_valid(self):
        return (self.a + self.b > self.c and
                self.a + self.c > self.b and
                self.b + self.c > self.a)
    def perimeter(self):
        return self.a + self.b + self.c
    def area(self):
        p = self.perimeter() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
    def display(self):
        print(f"Tam giác có cạnh: a={self.a}, b={self.b}, c={self.c}")
        if self.is_valid():
            print("  Chu vi:", self.perimeter())
            print("  Diện tích:", self.area())
        else:
            print("  Tam giác không hợp lệ!")
class RightTriangle(Triangle):
    def is_right(self):
        sides = sorted([self.a, self.b, self.c])
        return math.isclose(sides[0]**2 + sides[1]**2, sides[2]**2)
    def area(self):
        if self.is_right():
            sides = sorted([self.a, self.b, self.c])
            return (sides[0] * sides[1]) / 2
        else:
            return super().area()
    def display(self):
        print(f"Tam giác vuông có cạnh: a={self.a}, b={self.b}, c={self.c}")
        if self.is_valid():
            print("  Là tam giác vuông?", self.is_right())
            print("  Chu vi:", self.perimeter())
            print("  Diện tích:", self.area())
        else:
            print("  Tam giác không hợp lệ!")
class IsoscelesTriangle(Triangle):
    def is_isosceles(self):
        return self.a == self.b or self.a == self.c or self.b == self.c
    def display(self):
        print(f"Tam giác cân có cạnh: a={self.a}, b={self.b}, c={self.c}")
        if self.is_valid():
            print("  Là tam giác cân?", self.is_isosceles())
            print("  Chu vi:", self.perimeter())
            print("  Diện tích:", self.area())
        else:
            print("  Tam giác không hợp lệ!")
class EquilateralTriangle(IsoscelesTriangle):
    def is_equilateral(self):
        return self.a == self.b == self.c
    def area(self):
        if self.is_equilateral():
            return (math.sqrt(3) / 4) * self.a * self.a
        else:
            return super().area()
    def display(self):
        print(f"Tam giác đều có cạnh: a={self.a}, b={self.b}, c={self.c}")
        if self.is_valid():
            print("  Là tam giác đều?", self.is_equilateral())
            print("  Chu vi:", self.perimeter())
            print("  Diện tích:", self.area())
        else:
            print("  Tam giác không hợp lệ!")
```
